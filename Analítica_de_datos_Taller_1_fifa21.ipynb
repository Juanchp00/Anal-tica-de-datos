##Regresión lineal en sklearn##
**sklearn.linear_model.LinearRegression**
Es un algoritmo de aprendizaje automático basado en el aprendizaje supervisado, que modela un valor de predicción basado en variables independientes, usualmente  determina la relación entre las variables y la previsión.

La Regresión lineal se ajusta a un modelo lineal con coeficientes $w = (w1,…, wp)$ para minimizar la suma residual de cuadrados entre los objetivos observados en el conjunto de datos y los objetivos predichos por la aproximación lineal.

El codigo para realizar una regresión lineal es

***class sklearn.linear_model.LinearRegression( * , fit_intercept = True , normalize = False , copy_X = True , n_jobs = None , positive = False )***

Cuyos parámetros representan: 

* **fit_intercept bool, default=True**

  Sirve para calcular la intersección para este modelo si esta en **True**. Sin embargo, se establece en **False**, si no se utilizará ninguna intersección en los cálculos, por lo que se asume que los datos esten centralizados.

* **normalize: bool, default=False**

  Si fit_intercept esta en false el parámetro es ignorado. Sin embargo si fit_intecept es True, los regresores X se normalizan antes de la regresión mediante la resta entre la media y la divición por la norma $l_2$.

* **copy_X: bool, default=True**

  Si el parámetro es **True**, se copia X; si es **false**, X puede sobrescribirse.

* **n_jobs: int, default= None**

  Este parámetro va a proporcionar el número de trabajos que se utilizarán para el cálculo. Estas solo acelerarán para n_targets> 1 y crearán suficientes problemas grandes. Si el parametro esta en **None** significa 1, si el parámetro está en **$-1$** se utilizaran todos los procesadores.

* **positive: bool, default=False**

  Si el parámetro es **True**, forza  a los coeficientes a ser positivos, sin embargo, solo es copatible con matrices densas.


Por otro lado, sus atributos determinan como

* **coef_: array of shape (n_features, ) or (n_targets, n_features)**

  Se refiere a los coeficientes estimados para el problema de regresión lineal. Si se pasan varios objetivos durante el ajuste, se tendría una matriz de dos dimensiones de forma (n_targets, n_features), mientras que si solo se pasa un objetivo, sería una matriz de una dimensión de longitud n_features.

* **rank_: int**

  Se refiere al rango de matriz X. Solo disponible cuando la matriz es densa.

* **singular_: array of shape (min(X, y),)**

  Se refiere a los valores singulares de X. Solo disponible cuando la matriz es densa.

* **intercept_: float or array of shape (n_targets,)**

  Se refiere al término independiente en el modelo lineal. Se establece en 0.0 si .fit_intercept = False.

\\

##Regresión lineal Ridge en sklearn##
**sklearn.linear_model.Ridge**
es un tipo de regresión lineal utilizado para resolver un modelo de regresión donde la función de pérdida es la función de mínimos cuadrados lineales y la regularización es $L_2$. Este, reduce los coeficientes para las variables de entrada que no contribuyen mucho en la predicción.

El codigo para realizar una regresión lineal Ridge es

***class sklearn.linear_model.Ridge(alpha=1.0, *, fit_intercept=True, normalize=False, copy_X=True, max_iter=None, tol=0.001, solver='auto', random_state=None)***

Cuyos parámetros representan: 

* **alpha: {float, ndarray of shape (n_targets,)}, default=1.0**

 Da la Fuerza de regularización, este debe ser un numero flotante positivo. La regularización va a mejorar el condicionamiento del problema reduciendo la varianza de las estimaciones. **Un valor del parámetro más grande** especifican una regularización más fuerte.

* **fit_intercept: bool, default=True**

 Sirve para calcular la intersección para este modelo si esta en **True**. Sin embargo, se establece en **False**, si no se utilizará ninguna intersección en los cálculos, por lo que se asume que los datos esten centralizados.

* **normalize: bool, default=False**

  Si fit_intercept esta en false el parámetro es ignorado. Sin embargo si fit_intecept es True, los regresores X se normalizan antes de la regresión mediante la resta entre la media y la divición por la norma $l_2$.

* **copy_X: bool, default=True**

  Si el parámetro es **True**, se copia X; si es **false**, X puede sobrescribirse.

* **max_iter: int, default=None**

  Representa el número máximo de iteraciones para el resolutor del gradiente conjugado. Para los resolutores 'sparse_cg' y 'lsqr', scipy.sparse.linalg determina el valor predeterminado. Para el resolutor 'sag', el valor predeterminado es 1000.

* **tol: float, default=1e-3**

  Da la precisión de la solución.

* **solver: {‘auto’, ‘svd’, ‘cholesky’, ‘lsqr’, ‘sparse_cg’, ‘sag’, ‘saga’}, default=’auto’**

  Representa el resolutor para usar en las rutinas computacionales:

  * 'auto': elige el resolutor automáticamente según el tipo de datos.

  * 'svd': Usa una descomposición de valores singulares de X para calcular los coeficientes de cresta. Es más estable para matrices singulares que 'cholesky'.

  * 'cholesky': Usa la función estándar scipy.linalg.solve para obtener una solución de forma cerrada.

  * 'sparse_cg': Usa el resolutor de gradiente conjugado que se encuentra en scipy.sparse.linalg.cg. Como algoritmo iterativo, es más apropiado que 'cholesky' para datos a gran escala.

  * 'lsqr': Usa la rutina dedicada de mínimos cuadrados regularizados scipy.sparse.linalg.lsqr. Es el más rápido y utiliza un procedimiento iterativo.

  * 'sag': Usa un descenso de gradiente medio estocástico, y 'saga' usa su versión mejorada e imparcial llamada SAGA. Ambos métodos también utilizan un procedimiento iterativo y, a menudo, son más rápidos que otros resolutores cuando tanto n_samples como n_features sean grandes. La convergencia rápida 'sag' y 'saga' solo está garantizada en entidades con aproximadamente la misma escala. Puede preprocesar los datos con un escalador de sklearn.preprocessing.

Los últimos cinco resolutores admiten datos densos y escasos. Sin embargo, solo 'sag' y 'sparse_cg' admiten entrada escasa cuando fit_interceptes True.

* **random_state: int, RandomState instance, default=None**

  Se usa cuando solver== 'sag' o 'saga' para mezclar los datos.


Por otro lado, sus atributos determinan como

* **coef_: ndarray of shape (n_features,) or (n_targets, n_features)**

 Se refiere al peso de el o los vectores.

* **intercept_: float or ndarray of shape (n_targets,)**

  Se refiere al término independiente en función de decisión. Se establecer en 0.0 si .fit_intercept = False

* **n_iter_: None or ndarray of shape (n_targets,)**

  Representa el número real de iteraciones para cada objetivo. Disponible solo para solucionadores de sag y lsqr. Otros solucionadores devolverán "None".

\\

##Regresion lineal de Lazo de sklearn##
**sklearn.linear_model.Lasso**
es una modificación de la regresión lineal,la función de pérdida se modifica para minimizar la complejidad del modelo limitando la suma de los valores absolutos de los coeficientes del modelo, es decir, la norma $L_1$. El uso de una restricción de norma l1 obliga a algunos valores de ponderación a valer cero para permitir que otros coeficientes tomen valores distintos de cero, con el fin de reducir los coeficientes para aquellas variables de entrada que no contribuyen mucho a la tarea de predicción.

*Función objetivo*

$(1 / (2 * n_{samples})) * ||y - Xw||^2_2 + alpha * ||w||_1$

El modelo Lasso optimiza la misma función objetivo que el Elastic Net sin penalización $L_2$.


El codigo para realizar una regresión lineal de lasso es

**class sklearn.linear_model.Lasso( alpha = 1.0 , * , fit_intercept = True , normalize = False , precompute = False , copy_X = True , max_iter = 1000 , tol = 0.0001 , warm_start = False , positivo = False , random_state = Ninguno , selección = 'cíclico' )**

Cuyos parámetros definen:

* **alpha:float, default=1.0**

  Representa a la constante que multiplica el término L1. El valor predeterminado es 1.0 y es equivalente a un mínimo cuadrado ordinario, resuelto por el objeto. Por razones numéricas, no se recomienda su uso con el objeto.

* **fit_intercept:bool, default=True**

  Sirve para calcular la intersección para este modelo si esta en **True**. Sin embargo, se establece en **False**, si no se utilizará ninguna intersección en los cálculos, por lo que se asume que los datos esten centralizados.

* **normalize: bool, default=False**

  Si fit_intercept esta en false el parámetro es ignorado. Sin embargo si fit_intecept es True, los regresores X se normalizan antes de la regresión mediante la resta entre la media y la divición por la norma $l_2$.

* **precompute: ‘auto’, bool or array-like of shape (n_features, n_features), default=False**

  El parametro es **True** si se debe utilizar una matriz de Gram precalculada para acelerar los cálculos y dejar en 'auto' para decidir. La matriz de Gram también se puede pasar como argumento. Para una entrada escasa, esta opción siempre es **True** para preservar la escasez.

* **copy_X:bool, default=True**

  Si el parámetro es **True**, se copia X; si es **false**, X puede sobrescribirse.

* **max_iter: int, default=1000**

  Representa el número máximo de iteraciones.

* **tol: float, default=1e-4**

  Representa la tolerancia para la optimización: el código de optimización comprueba la optimización del espacio dual y continúa hasta que sea más pequeño que tol.

* **warm_start: bool, default=False**

  Cuando se establece en **True**, reutiliza la solución de la llamada anterior para que encaje como inicialización; de lo contrario, simplemente borra la solución anterior.

* **positive: bool, default=False**

  Si el parámetro es **True**, forza  a los coeficientes a ser positivos.

* **random_state: int, RandomState instance, default=None**

  Representa el generador de números aleatorios, selecciona una característica aleatoria para actualizar.

* **selection: {‘cyclic’, ‘random’}, default=’cyclic’**

  Si se establece en **random**, un coeficiente aleatorio se actualiza en cada iteración, si se establece **cyclic** recorre las características secuencialmente de forma predeterminada. Establecerlo en **random** a menudo conduce a una convergencia significativamente más rápida, especialmente cuando tol es mayor que 1e-4.


Por otro lado, sus atributos representan

* **coef_: ndarray of shape (n_features,) or (n_targets, n_features)**

  El vector de parametros (w en la función de costos)

* **dual_gap_: float or ndarray of shape (n_targets,)**

  Dado el parámetro alfa, los espacios dobles al final de la optimización, tienen la misma forma que cada observación de y.

* **sparse_coef_: sparse matrix of shape (n_features, 1) or (n_targets, n_features)**

  Representación escasa del ajustado coef_.

* **intercept_: float or ndarray of shape (n_targets,)**

  Término independiente en función de decisión.

* **n_iter_: int or list of int**

  Número de iteraciones ejecutadas por el resolutor de descenso de coordenadas para alcanzar la tolerancia especificada.

\\

##Regresion lineal ElasticNet de sklearn##

**sklearn.linear_model.ElasticNet**
es una modificación de la regresión lineal que combina las propiedades de la regresión Ridge y Lasso, penalizando el modelo utilizando tanto la norma $L_2$ como la norma $L_1$.
La regresión es útil cuando hay varias funciones correlacionadas y La diferencia con la regresión lineal de Lazo radica en que es probable que Lasso elija una de estas características al azar, mientras que ElasticNet probablemente elija ambas a la vez, tal combinación permite aprender un modelo disperso donde pocos de los pesos son distintos de cero como el método de regularización de Lasso, mientras se mantienen las propiedades de regularización del método de regularización de Ridge.

Esta regresión minimiza la función objetivo

$$1 / (2 * n_samples) * ||y - Xw||^2_2
+ alpha * l1_ratio * ||w||_1
+ 0.5 * alpha * (1 - l_{1_{ratio}}) * ||w||^2_2$$

Además, si está interesado en controlar la penalización L1 y L2 por separado, equivale a

$$a * L1 + b * L2$$

donde

$$alpha = a + b$$
y
$$l_{1_{ratio}} = a / (a + b)$$

El parámetro $l_{1_{ratio}}$ corresponde a alpha en el paquete glmnet R, mientras que alpha corresponde al parámetro lambda en glmnet. 

El codigo para realizar una regresión lineal ElasticNet es

**class sklearn.linear_model.ElasticNet(alpha=1.0, *, l1_ratio=0.5, fit_intercept=True, normalize=False, precompute=False, max_iter=1000, copy_X=True, tol=0.0001, warm_start=False, positive=False, random_state=None, selection='cyclic')**

Cuyos parámetros definen:

* **alpha: float, default=1.0**
  
  Representa a la constante que multiplica los términos de penalización. El valor predeterminado es 1.0.


* **l1_ratio: float, default=0.5**

  Representa el parámetro de mezcla de ElasticNet, para $0 <= l_{1_{ratio}} <= 1$. Con $l_{1_{ratio}} = 0$ la pensalizacion es de $L_2$, Para $l_{1_{ratio}} = 1$ la penalización es de $L_1$, y para $0 < l_{1_{ratio}} < 1$, la penalización es combinacion de $L_1$ y $L_2$.

* **fit_intercept: bool, default=True**

  Sirve para calcular la intersección para este modelo si esta en **True**. Sin embargo, se establece en **False**, si no se utilizará ninguna intersección en los cálculos, por lo que se asume que los datos esten centralizados

* **normalize: bool, default=False**

  Si fit_intercept esta en false el parámetro es ignorado. Sin embargo si fit_intecept es **True**, los regresores X se normalizan antes de la regresión mediante la resta entre la media y la divición por la norma $l_2$.


* **precompute: bool or array-like of shape (n_features, n_features), default=False**

  El parametro es **True** si se debe utilizar una matriz de Gram precalculada para acelerar los cálculos y dejar en 'auto' para decidir. La matriz de Gram también se puede pasar como argumento. Para una entrada escasa, esta opción siempre es **True** para preservar la escasez.

* **max_iter: int, default=1000**

  Representa el número máximo de iteraciones.

* **copy_X: bool, default=True**

  Si el parámetro es **True**, se copia X; si es **false**, X puede sobrescribirse.

* **tol: float, default=1e-4**

  Representa la tolerancia para la optimización: el código de optimización comprueba la optimización del espacio dual y continúa hasta que sea más pequeño que tol.

* **warm_start: bool, default=False**

  Cuando se establece en **True**, reutiliza la solución de la llamada anterior para que encaje como inicialización; de lo contrario, simplemente borra la solución anterior.

* **positive: bool, default=False**

  Si el parámetro es **True**, forza  a los coeficientes a ser positivos.

* **random_state: int, RandomState instance, default=None**

  Representa el generador de números aleatorios, selecciona una característica aleatoria para actualizar.

* **selection{‘cyclic’, ‘random’}, default=’cyclic’**
 
 Si se establece en **random**, un coeficiente aleatorio se actualiza en cada iteración, si se establece **cyclic** recorre las características secuencialmente de forma predeterminada. Establecerlo en **random** a menudo conduce a una convergencia significativamente más rápida, especialmente cuando tol es mayor que 1e-4.

##Regresion KernelRidge de sklearn##

**sklearn.kernel_ridge.KernelRidge**
Se trata de una regresión no lineal, el cual es más eficiente cuando se trabaja con datos complejos, puesto a que las regresiones lineales a menudo no son suficientes para captar algunas peculiaridades del problema, a lo que la no linealidad podría ser favorable.
Esta regresión combina la regresión Ridge y la clasificación por mínimos cuadrados lineales con regularización de norma $L_2$ con el kernel trick. Así, la regresión aprende una función lineal en el espacio inducido por el kernel respectivo y los datos, sin embargo,Kernels no lineales, va a corresponder a funciones no lineales en el espacio original. 


El codigo para realizar una regresión KernelRidge es

**class sklearn.kernel_ridge.KernelRidge(alpha=1, *, kernel='linear', gamma=None, degree=3, coef0=1, kernel_params=None)**

Cuyos parámetros definen:

* **alpha: float or array-like of shape (n_targets,), default=1.0**
  
  Da la Fuerza de regularización, este debe ser un numero flotante positivo. La regularización va a mejorar el condicionamiento del problema reduciendo la varianza de las estimaciones. Un valor del parámetro más grande especifican una regularización más fuerte.


* **kernel: string or callable, default=”linear”**

  Representa al mapeo de kernel utilizado internamente. Este parámetro se pasa directamente a pairwise_kernel. Si kernel es un "**string**", debe ser una de las métricas en pairwise.PAIRWISE_KERNEL_FUNCTIONS. Si kernel es "**precomputed**", se supone que X es una matriz kernel. Alternativamente, si kernel es una función **callable**, se llama en cada par de instancias (filas) y se registra el valor resultante. El callable debe tomar dos filas de X como entrada y devolver el valor del kernel correspondiente como un solo número. Esto significa que las llamadas desde sklearn.metrics.pairwiseno están permitidas, ya que operan en matrices, no en muestras individuales. En su lugar, se utiliza la cadena que identifica el kernel.


* **gamma: float, default=None**

  El parámetro gamma para los núcleos RBF, laplaciano, polinomial, chi2 exponencial y sigmoide, la interpretación del valor predeterminado se deja al kernel.

* **degree: float, default=3**

  Representa el grado del núcleo polinomial. Ignorado por otros núcleos.


* **coef0: float, default=1**

  Representa el coeficiente cero para núcleos polinomiales y sigmoides. Ignorado por otros núcleos.

* **kernel_params: mapping of string to any, default=None**

  Representa los parámetros adicionales (argumentos de palabras clave) para la función del kernel pasados ​​como objeto invocable.


Por otro lado, sus atributos representan

* **dual_coef_ndarray of shape (n_samples,) or (n_samples, n_targets)**

  Representación de vector o vectores de peso en el espacio kernel

* **X_fit_{ndarray, sparse matrix} of shape (n_samples, n_features)**

  Representa los datos de entrenamiento, que también son necesarios para la predicción. Si kernel es **“precomputed”**, esta es en cambio la matriz de entrenamiento precalculada, de forma (n_samples, n_samples).

```python
#%%   
#codificar estadísticas -> jugadores fifa ej 88+3
def code_stats(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
        #print(y.iloc[i])
        
        if y.iloc[i].find("+") > -1: # encontrar signo mas en str y casteo a flotante
            yc[i] = float(y.iloc[i][:y.iloc[i].find("+")])+float(y.iloc[i][y.iloc[i].find("+")+1:])
        else: yc[i] = float(y.iloc[i])
        #print(yc[i])
    return yc   

#codificar moneda ->
def code_euro(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
      #  print(i,y[i])
        if y[i][-1]=='M': yc[i] = float(y[i][1:-1])*10**6 #buscar M y reemplazar 10^6
        elif y[i][-1]=='K': yc[i] = float(y[i][1:-1])*10**3 # buscar K y reemplazar por 10^3
        else: yc[i] = float(y[i][1:])
        #print(yc[i])
    return yc 

def code_hits(y):
    yc = np.zeros(y.shape[0])
    for i in range(y.shape[0]):
      #  print(type,y[i])
        y[i]=str(y[i])
        
        if y[i][-1]=='M': yc[i] = float(y[i][0:-1])*10**6 
        elif y[i][-1]=='K': yc[i] = float(y[i][0:-1])*10**3 
        else: yc[i] = float(y[i][0:])
        #print(yc[i])
    return yc 
```

###**Clase de preproceso**


```python
#%% dummy transformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder

class dummy_fifa(BaseEstimator,TransformerMixin):
    #inicializacion de clase y varaibles
    def __init__(self, col_euro,col_stats, cat_usr): #constructor clase
        self.col_euro = col_euro #lista atributos tipo moneda
        self.col_stats = col_stats #lista atributos tipo estadistica
        self.cat_usr = cat_usr #lista de atributos categoricos

    def fit(self,X, *_):
        Xi = X.copy() #copiar dataset para no reemplazar original
        self.imputer_num = SimpleImputer(strategy="most_frequent") #crear imputador tipo modo
        self.a = Xi.columns[np.sum(Xi.isna())> 0] #encontrar columnas con datos faltantes
        #print(a)
        self.imputer_num.fit(Xi[self.a]) # ajustar imputador
        
        Xi[self.a] = self.imputer_num.transform(Xi[self.a]) #evaluar  sobre datos imputador
        #print('Cod Euros\n')
        for i in self.col_euro: #codificar tipo moneda
         #   print(i)
            Xi[i] = code_euro(np.array(Xi[i]))
        #print('Cod stats\n')    
        for i in self.col_stats: #codificar datos estadisticos
         #   print(i)
            Xi[i] = code_stats(Xi[i])
        
        for i in ['Hits']: 
         #   print(i)
            Xi[i] = code_hits(np.array(Xi[i]))
            
        
        #height, wieght corregir formato
        Xi['Height'].replace(regex=["'"], value='.',inplace=True) 
        Xi['Height'].replace(regex=['"'], value='',inplace=True)        
        for i in Xi.index:
            #print(float(Xi.loc[i,'Weight'][:-3]))
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])
        
        Xi['Height'] = Xi['Height'].astype('float64');#print(Xi['Height'].dtype)

        ##cambiamos fecha por flotante
        Xi['Joined'] = Xi['Joined'].replace(regex="Jan",value="01")
        Xi['Joined'] = Xi['Joined'].replace(regex="Feb",value="02")
        Xi['Joined'] = Xi['Joined'].replace(regex="Mar",value="03")
        Xi['Joined'] = Xi['Joined'].replace(regex="Apr",value="04")
        Xi['Joined'] = Xi['Joined'].replace(regex="May",value="05")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jun",value="06")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jul",value="07")
        Xi['Joined'] = Xi['Joined'].replace(regex="Aug",value="08")
        Xi['Joined'] = Xi['Joined'].replace(regex="Sep",value="09")
        Xi['Joined'] = Xi['Joined'].replace(regex="Oct",value="10")
        Xi['Joined'] = Xi['Joined'].replace(regex="Nov",value="11")
        Xi['Joined'] = Xi['Joined'].replace(regex="Dec",value="12")    
        Xi['Joined'] = Xi['Joined'].replace(regex=",",value=".")
        Xi['Joined'] = Xi['Joined'].replace(regex=" ",value="")
        Xi['Joined'] = Xi['Joined'].astype('float64')

        ##borramos caracter estrella
        Xi['W/F'] = Xi['W/F'].replace(regex=" ★",value="")
        Xi['W/F'] = Xi['W/F'].replace(regex="★",value="")
        Xi['W/F'] = Xi['W/F'].astype('float64')

        Xi['SM'] = Xi['SM'].replace(regex=" ★",value="")
        Xi['SM'] = Xi['SM'].replace(regex="★",value="")
        Xi['SM'] = Xi['SM'].astype('float64')

        Xi['IR'] = Xi['IR'].replace(regex=" ★",value="")
        Xi['IR'] = Xi['IR'].replace(regex="★",value="")
        Xi['IR'] = Xi['IR'].astype('float64')
        

        cat = [] #codificar variables categoricas con ordinal encoder
        for i in self.cat_usr.keys():
            cat = cat + [[*self.cat_usr.get(i)]]
        self.col_cat_usr = OrdinalEncoder(categories=cat)
        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.fit_transform(Xi[[*self.cat_usr.keys()]])
        
        #self.col_cat = Xi.columns[Xi.dtypes=='O']
        #self.cod = OrdinalEncoder()
        #self.cod.fit(Xi[self.col_cat])
        
        return self    

    def transform(self, X, *_):#funcion transformador
        Xi = X.copy()
        Xi[self.a] = self.imputer_num.transform(Xi[self.a])
        
        for i in self.col_euro:
         #   print(i)
            Xi[i] = code_euro(np.array(Xi[i]))
        #print('Cod stats\n')    
        for i in self.col_stats:
         #   print(i)
            #Xi[i] = code_hits(Xi[i])
           Xi[i] = code_stats(Xi[i])

        for i in ['Hits']: 
         #   print(i)
            Xi[i] = code_hits(np.array(Xi[i]))
        
        #height, wieght
        Xi['Height'].replace(regex=["'"], value='.',inplace=True)
        Xi['Height'].replace(regex=['"'], value='',inplace=True)        
        for i in Xi.index:
            #print(float(Xi.loc[i,'Weight'][:-3]))
            Xi.loc[i,'Weight'] = float(Xi.loc[i,'Weight'][:-3])
            Xi.loc[i,'Height'] = float(Xi.loc[i,'Height'])
        
        Xi['Height'] = Xi['Height'].astype('float64')
        Xi['Weight'] = Xi['Weight'].astype('float64')

        Xi['Joined'] = Xi['Joined'].replace(regex="Jan",value="01")
        Xi['Joined'] = Xi['Joined'].replace(regex="Feb",value="02")
        Xi['Joined'] = Xi['Joined'].replace(regex="Mar",value="03")
        Xi['Joined'] = Xi['Joined'].replace(regex="Apr",value="04")
        Xi['Joined'] = Xi['Joined'].replace(regex="May",value="05")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jun",value="06")
        Xi['Joined'] = Xi['Joined'].replace(regex="Jul",value="07")
        Xi['Joined'] = Xi['Joined'].replace(regex="Aug",value="08")
        Xi['Joined'] = Xi['Joined'].replace(regex="Sep",value="09")
        Xi['Joined'] = Xi['Joined'].replace(regex="Oct",value="10")
        Xi['Joined'] = Xi['Joined'].replace(regex="Nov",value="11")
        Xi['Joined'] = Xi['Joined'].replace(regex="Dec",value="12")    
        Xi['Joined'] = Xi['Joined'].replace(regex=",",value=".")
        Xi['Joined'] = Xi['Joined'].replace(regex=" ",value="")
        Xi['Joined'] = Xi['Joined'].astype('float64')

        Xi['W/F'] = Xi['W/F'].replace(regex=" ★",value="")
        Xi['W/F'] = Xi['W/F'].replace(regex="★",value="")
        Xi['W/F'] = Xi['W/F'].astype('float64')

        Xi['SM'] = Xi['SM'].replace(regex=" ★",value="")
        Xi['SM'] = Xi['SM'].replace(regex="★",value="")
        Xi['SM'] = Xi['SM'].astype('float64')

        Xi['IR'] = Xi['IR'].replace(regex=" ★",value="")
        Xi['IR'] = Xi['IR'].replace(regex="★",value="")
        Xi['IR'] = Xi['IR'].astype('float64')
        
        Xi[[*self.cat_usr.keys()]] =self.col_cat_usr.transform(Xi[[*self.cat_usr.keys()]])
        #Xi[self.col_cat]= self.cod.transform(Xi[self.col_cat])
        return Xi
    
    def fit_transform(self,X,*_):
        self.fit(X)
        return self.transform(X)
```
